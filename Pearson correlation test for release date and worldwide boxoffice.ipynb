```python
# Perform the Pearson correlation test to evaluate the linear correlation between the two continuous variables release date and worldwide boxoffice.

import psycopg2
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Function for database connection
def connectDB():
    conn = psycopg2.connect(
        dbname="Movie Industry Database",
        user="postgres",
        password="Quickborn123",
        host="localhost",
        port="5432")
    return conn

# Query to select release_date and worldwide_boxoffice
def get_data(conn):
    df = pd.read_sql("""SELECT release_date, worldwide_boxoffice
                        FROM movies""", conn)
    return df

# Function to calculate Pearson correlation
def calculate_pearson_correlation(df):
    correlation_coefficient, p_value = stats.pearsonr(df['release_date'], df['worldwide_boxoffice'])
    return correlation_coefficient, p_value

# Establish a connection to the database
conn = connectDB()
print("Database connection established.")

# Retrieve data
df = get_data(conn)
print(df)

# Calculate Pearson correlation coefficient and p-value
correlation_coefficient, p_value = calculate_pearson_correlation(df)
print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

# Create a scatter plot
plt.scatter(df['release_date'], df['worldwide_boxoffice'])
plt.xlabel('Release Date')
plt.ylabel('Worldwide Box Office')
plt.title('Correlation between Release Date and Worldwide Box Office')
plt.show()

# Close the database connection
conn.close()
print("Database connection closed.")

```
